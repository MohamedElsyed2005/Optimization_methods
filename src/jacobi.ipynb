```python
import numpy as np 
from Unconstrained.Jacobi import jacobi
```

```python
A = [[10, 2, -3],
     [2, 8, 1],
     [3, -1, 15]]

b = [5, 6, 12]
points = [0.5, 1, 1] 
solver = jacobi(A = A, b = b, max_iteration = 8, points = points)
solver.fit()
```
