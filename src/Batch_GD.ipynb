```python
import numpy as np 
from Unconstrained.GradientDescent import Batch_GD
```

```python
np.random.seed(42)
x = 2 * np.random.rand(100,1)
y =  4 + 3 * x + np.random.randn(100,1)

model = Batch_GD(x, y, learning_rate=0.1, tolerance=1e-6, max_iters=1000)
model.fit()

print("Final Parameters:", model.parameters)
print("Prediction for 6:", model.predict(np.array([[6]])))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color="blue", label="Data")
plt.plot(x, model.predict(x), color="red", linewidth=2, label="Model Prediction")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
