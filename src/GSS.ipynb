```python
import numpy as np
import math
from Unconstrained.Golden_Section_Search import GSS
```

```python
def target_function(theta):
    return 4 * np.sin(theta) * (1 + np.cos(theta))

gss = GSS(func = target_function, a = 0, b = np.pi / 2, find_max = True)
gss.fit()
```

```python
def f(x):
    return (math.exp(-x) - np.cos(x))

gss = GSS(func = f, a = 0, b = 1)
gss.fit()
```

```python
def f(x):
    return (2*x + 2*np.cos(x))

gss = GSS(func = f , a = 0, b = np.pi)
gss.fit()
```

```python
def f(x):
    return ((2 * np.sin(x)) - (x**2 / 10))

gss =  GSS(func = f, a = 0 , b = 4, find_max = True)
gss.fit()
```
